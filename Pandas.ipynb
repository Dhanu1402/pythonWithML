Creating a Pandas dataframe

```python
# impporting the boston house price data
from sklearn.datasets import fetch_california_housing
```

```python
california_datasets = fetch_california_housing()
```

Manipulating a DataFrame

```python
# adding a column in a 
california_df['Price'] = california_datasets.target
```

```python
california_df.head()
```

```python
# removing a row --> when deleting a row then axis = 0 and while deleting a column axis = 1
california_df.drop(index = 0, axis = 0)
```

```python
# drop a column 
# to delete rows and columns permanent store the data into new variable 
california_df.drop(columns = 'Latitude', axis = 1)
```

```python
# locating a row using index value
california_df.iloc[2]
```

```python
# locating a particular column 
print(california_df.iloc[:, 0])  # first column
print(california_df.iloc[:, 1])  # second column
print(california_df.iloc[:, 2])  # third column
print(california_df.iloc[:, -1]) # last column
```

Correlation (relation between columns)
