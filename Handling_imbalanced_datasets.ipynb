Imbalanced Dataset:
A dataset with an unequal class distribution

```python
# importing the dependencies
import numpy as np
import pandas as pd
```

```python
# loading tha dataset to pandas DataFrame
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# distribution of two classes
credit_card_data['Class'].value_counts()
```

This is Higly Imbalanced Dataset

0 --> Legit Transactions

1 --> Fraudulent Transactions

```python
# separating the legit and fraudulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-sampling

Build a sample dataset containing sumilar distribution of Legit & Fraudulent Transactions

Number of Fraudulent Transactions --> 73

```python
legit_sample = legit.sample(n = 73)
```

```python
print(legit_sample.shape)
```

Concatenate the Two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0) 
# here axis = 0 means we need to concatenate row vise
```
